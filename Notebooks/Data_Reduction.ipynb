üëâDefining Paths for train, validation and test data

```python
train_image = '../Data/Split/train/data/'
train_mask = '../Data/Split/train/mask/'

val_image = '../Data/Split/val/data/'
val_mask = '../Data/Split/val/mask/'

test_image = '../Data/Testing/data/'
```

 

üëâ for TRAIN IMAGES

```python
for img in os.listdir(train_image):
    data = np.load(train_image + img)
    if data.dtype == np.float64:
        data = data.astype(np.float32)
    np.save(train_image + img, data)
```

üëâ for TRAIN MASKS

```python
for img in os.listdir(train_mask):
    data = np.load(train_mask + img)
    if data.dtype == np.float64:
        data = data.astype(np.float32)
    np.save(train_mask + img, data)
```
